```python
import numpy as np
import matplotlib.tri as mptri
import matplotlib.pyplot as plt
from netgen.geom2d import *
from ngsolve import *
import os, meshio
plt.rc("text", usetex=True)
meshDir = os.path.join(os.getcwd(), "meshes2D")
```

```python
def generateNetgenMesh(hval):
    geo = SplineGeometry()

    rect = geo.AddRectangle((-0.1, -0.1), (0.1, 0.1), leftdomain=1, rightdomain=0, bc="outer")
    circ = geo.AddCircle((0, 0), 0.025, leftdomain=0, rightdomain=1, bc = "inner")

    msh = geo.GenerateMesh(maxh=hval)

    points = []
    for pt in msh.Points():
        points.append([*pt])

    points = np.array(points)

    points = points[:, :2]

    el = []
    for elem in msh.Elements2D():
        el.append([int(str(i)) for i in elem.points])

    el = np.array(el, int)
    return msh, points, el-1
```

```python
num_refinements = 3
fig, axs = plt.subplots(1,3,figsize=(15,5))
for i in range(num_refinements):
    msh, pts, els = generateNetgenMesh(0.02/(i+1))    
    axs[i].triplot(pts[:, 0], pts[:, 1], els)
    axs[i].set_title("ne = ${}$".format(els.shape[0]), fontsize=16)
    filename_mesh = "mesh_{}.msh".format(i+1)
    msh.Export(os.path.join(meshDir, filename_mesh), "Gmsh2 Format")
    fil = meshio.read(os.path.join(meshDir, filename_mesh).format(i+1))
    for cell in fil.cells:
        if cell.type == "triangle":
            cell_data = cell.data
        
    mMesh = meshio.Mesh(points=fil.points[:, :2], cells = {"triangle": cell_data} )
    meshio.xdmf.write(os.path.join(meshDir, "mesh_{}.xdmf".format(i+1)), mMesh)
    
```

```python
from dolfin import *
from vtkplotter.dolfin import plot as vplot
for i in range(num_refinements):
    dolfinMesh = Mesh()
    XDMFFile(os.path.join(meshDir, "mesh_{}.xdmf".format(i+1))).read(dolfinMesh)
    plt.figure()
    plot(dolfinMesh)
#     vplot(dolfinMesh)
```
