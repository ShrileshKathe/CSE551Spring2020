```python
from dolfin import *
import numpy as np
import os
import matplotlib.pyplot as plt
from matplotlib import cm
plt.rc("text", usetex=True)
```

```python
msh = Mesh()
XDMFFile(os.path.join(os.getcwd(), "meshes2D","mesh_3.xdmf")).read(msh)
```

```python
plt.figure(figsize=(8,8))
plot(msh)
```

```python
leftpt = CompiledSubDomain("near(x[0], -0.1) && near(x[1], -0.1) && on_boundary")
right = CompiledSubDomain("near(x[0], 0.1) && on_boundary")
left = CompiledSubDomain("near(x[0], -0.1) && on_boundary")
top = CompiledSubDomain("near(x[1], 0.1) && on_boundary")
bottom = CompiledSubDomain("near(x[1], -0.1) && on_boundary")
circ = CompiledSubDomain("near((pow(x[0],2) + pow(x[1],2))/pow(0.025, 2), 1, 0.1 ) && on_boundary")



bdryMark = MeshFunction("size_t", msh, msh.geometry().dim() - 1)
bdryMark.set_all(0)

right.mark(bdryMark, 1)
top.mark(bdryMark, 2)
left.mark(bdryMark, 3)
bottom.mark(bdryMark, 4)

V = FunctionSpace(msh, "CG", 2)
Vq = VectorFunctionSpace(msh, "DG", 1)
u_ = TrialFunction(V)
du = TestFunction(V)
u = Function(V)
ds = Measure("ds", subdomain_data = bdryMark)

bcs = DirichletBC(V, Constant(0), leftpt) #conditioning

alph1 = Constant(10.)
alph2 = Constant(-10.)

a = 1*inner(grad(u_), grad(du))*dx
L = inner(alph1, du)*ds(2) + inner(alph2, du)*ds(4)
solve(a==L, u, [])
```

```python
plt.figure(figsize=(10,8))
p1 = plot(u, cmap=cm.jet)
plt.colorbar(p1)
plt.title("Temperature: $T(x,y)$", fontsize=18)
```

```python
q = project(grad(u), Vq)
```

```python
plt.figure(figsize=(10,8))
q1 = plot(-q[0], cmap=cm.jet)
plt.colorbar(q1)
plt.title(r"Flux: $q_1 = -\frac{\partial T}{\partial {x_1}}$", fontsize=18)
```

```python
plt.figure(figsize=(10,8))
q2 = plot(-q[1], cmap=cm.jet)
plt.colorbar(q2)
plt.title(r"Flux: $q_2 = -\frac{\partial T}{\partial {x_2}}$", fontsize=18)
```
